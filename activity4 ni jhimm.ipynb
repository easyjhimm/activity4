<h1>2. What is the impact of light conditions on accident severity?</h1>

```python
light_severity = accidents.groupby('Light_Conditions')['Accident_Severity'].value_counts()
```

```python
light_severity
```

<h1>Insight No. 2</h1>
<h2>Most accidents occur in daylight with 419,045 slight, 60,759 serious, and 5,076 fatal accidents reported</h2>

```python
light_conditions = [
    "Darkness - lighting unknown", "Darkness - lights lit", "Darkness - lights unlit",
    "Darkness - no lighting", "Daylight"
]

slight = [5622, 108345, 2138, 28651, 419045]
serious = [794, 19130, 360, 7174, 60759]
fatal = [68, 1860, 45, 1612, 5076]

df = pd.DataFrame({
    "Light Conditions": light_conditions,
    "Slight": slight,
    "Serious": serious,
    "Fatal": fatal
})

df.set_index("Light Conditions", inplace=True)


fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.6
colors = ["#ffcc00", "#ff6600", "#cc0000"]  

bottom = np.zeros(len(light_conditions))
for severity, color in zip(["Slight", "Serious", "Fatal"], colors):
    ax.bar(df.index, df[severity], label=severity, color=color, bottom=bottom, width=bar_width)
    bottom += df[severity]

ax.set_ylabel("Number of Accidents")
ax.set_xlabel("Light Conditions")
ax.set_title("Impact of Light Conditions on Accident Severity")
ax.legend(title="Accident Severity")
plt.xticks(rotation=25)
plt.show()
```
